```python
import pandas as pd

# Load both files
meta = pd.read_csv("metadata_1939.txt", sep="\t")
run_table = pd.read_csv("SraRunTable.csv")

# Clean all column names
meta.columns = meta.columns.str.strip().str.lower()
run_table.columns = run_table.columns.str.strip()

# Add full anonymized_name format: 1939.XXXXX
run_table["anonymized_name_full"] = "1939." + run_table["anonymized_name"].astype(str)

# Convert all potential metadata ID columns to string
for col in ["sample_name", "orig_name", "anonymized_name", "host_subject_id"]:
    if col in meta.columns:
        meta[col] = meta[col].astype(str)

# Candidate ID columns to try join on
candidates = ["sample_name", "orig_name", "anonymized_name", "host_subject_id"]
best_match = None
best_count = -1
best_col = None
best_diag_col = None

# Try merging using each candidate ID column
for col in candidates:
    merged = run_table.merge(meta[[col, "diagnosis"]], left_on="anonymized_name_full", right_on=col, how="left")

    # Diagnosis column might be renamed to diagnosis_y if duplicate exists
    diag_col = "diagnosis_y" if "diagnosis_y" in merged.columns else "diagnosis"
    count = merged[diag_col].notna().sum()

    if count > best_count:
        best_match = merged.copy()
        best_count = count
        best_col = col
        best_diag_col = diag_col

# Rename diagnosis column to standard name
best_match = best_match.rename(columns={best_diag_col: "diagnosis"})

# Show preview
print(f"âœ… Best match using '{best_col}' â†’ {best_count} samples with diagnosis found.\n")
print("Diagnosis values:", best_match["diagnosis"].dropna().unique())
print("\nPreview:\n", best_match[["Run", "anonymized_name_full", best_col, "diagnosis"]].dropna().head())

# Save for next step
best_match.to_csv("run_table_with_diagnosis.csv", index=False)
print("\nðŸ“ Saved as: run_table_with_diagnosis.csv")
```

```python

import pandas as pd

# Load onehot vectors
vectors = pd.read_csv("onehot_vectors.csv")
vectors.columns = vectors.columns.str.strip()
if "graph_id" not in vectors.columns:
    vectors.rename(columns={vectors.columns[0]: "graph_id"}, inplace=True)

# Clean graph_id to match run table (remove _raw suffix)
vectors["clean_id"] = vectors["graph_id"].str.replace("_raw", "", regex=False)

# Load run-to-diagnosis mapping
run_diag = pd.read_csv("run_table_with_diagnosis.csv")
run_diag.columns = run_diag.columns.str.strip()

# Assign binary label
def assign_label(d):
    d = str(d).lower()
    if d in {"cd", "uc", "ic", "ibd"}:
        return 1
    elif d in {"control", "no", "healthy"}:
        return 0
    else:
        return None

run_diag["label"] = run_diag["diagnosis"].apply(assign_label)

# Merge vectors with labels
merged = vectors.merge(run_diag[["Run", "label"]], left_on="clean_id", right_on="Run", how="left")
final = merged.dropna(subset=["label"])
final["label"] = final["label"].astype(int)

# Save final dataset
final.to_csv("labeled_vectors.csv", index=False)
print("âœ… Final labeled_vectors.csv saved!")
print("ðŸ“Š Label distribution:", final["label"].value_counts().to_dict())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Load dataset
df = pd.read_csv("labeled_vectors.csv")

# Show label distribution
print("ðŸ“Š Full Dataset Label Distribution:")
print(df["label"].value_counts())

# Ensure we only keep numeric columns for features
# First drop non-feature columns, then select only numeric columns
X = df.drop(columns=["graph_id", "clean_id", "label"], errors='ignore').select_dtypes(include=['number'])
y = df["label"]

# Verify we have the right columns
print("\nðŸ” Features being used:")
print(X.columns.tolist())

# Split: 80% train, 20% test (stratified to preserve class ratio)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Show split results
print("\nâœ… Train set size:", len(X_train))
print("   Label distribution:", y_train.value_counts().to_dict())

print("\nâœ… Test set size:", len(X_test))
print("   Label distribution:", y_test.value_counts().to_dict())

# Apply SMOTE only to the training set
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Confirm resampling
print("\nâœ… SMOTE applied successfully!")
print("Resampled train set shape:", X_train_res.shape)
print("Resampled label counts:", pd.Series(y_train_res).value_counts().to_dict())

# Original test set remains unchanged
print("\nTest set shape remains:", X_test.shape)
```
